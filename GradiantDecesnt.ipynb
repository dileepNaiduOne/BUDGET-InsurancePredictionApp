```python
import pandas as pd
import numpy as np
import sklearn
```

```python
X = np.array([[2, 5, 7]])
Y = np.array([5, 12, 15])

B0 = 1 #np.random.randint(100)
B1 = 3 #np.random.randint(100)

learning_rate = 0.01

def Gradient(B0, B1, X, Y, learning_rate, iterations):

    for i in range(iterations):
        print("#"*80)
        print(f"STEP - {i+1}".center(80, "-"))
        print("#"*80)
        
        y_pred = B0 + B1*X
        print(f"Y_Pred --> {y_pred}")
    
        yi_y_pred = Y - y_pred
        print(f"Y_i - Y_Pred --> {yi_y_pred}")
    
        sse = 0.5*(yi_y_pred**2).sum()
        print(f"SSE --> {sse}")
    
        # Update B0 & B1
        dB0 = -(yi_y_pred).sum()
        print(f"Partial Derivative of B0 --> {dB0}")
        
        new_B0 = B0 - learning_rate * dB0
        print(f"New B0 --> {new_B0}")
        B0 = new_B0
    
        dB1 = -(X*(yi_y_pred)).sum()
        print(f"Partial Derivative of B1 --> {dB1}")
        
        new_B1 = B1 - learning_rate * dB1
        print(f"New B1 --> {new_B1}")
        B1 = new_B1
        print("-"*80)
        print()
        print()
        print()

Gradient(B0, B1, X, Y, learning_rate, 10)
```
